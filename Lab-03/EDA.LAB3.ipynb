```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['gender_label'] = le.fit_transform(df['gender'])
df[['gender', 'gender_label']].head()
```

```python
df_onehot = pd.get_dummies(df, columns=['marital_status'])
df_onehot.head()
```

```python
df['education_level'].unique()
```

```python
df['education_level'] = df['education_level'].astype(str)
df['education_level'] = df['education_level'].str.strip()
df['education_level'] = df['education_level'].str.title()
```

```python
print(df['education_level'].unique())
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
order = [['High School', 'Diploma', 'Bachelor', 'Master', 'Phd']]
```

```python
oe = OrdinalEncoder(
    categories=order,
    handle_unknown='use_encoded_value',
    unknown_value=-1
)
```

```python
df['education_ordinal'] = oe.fit_transform(df[['education_level']])
```

```python
print(df[['education_level', 'education_ordinal']].drop_duplicates())
```

```python
job_role_mean = df.groupby('job_role')['churn'].mean()
df['job_role_target'] = df['job_role'].map(job_role_mean)
df[['job_role', 'job_role_target']].head()
```

```python
freq_map = df['payment_method'].value_counts()
df['payment_method_freq'] = df['payment_method'].map(freq_map)
df[['payment_method', 'payment_method_freq']].head()
```
