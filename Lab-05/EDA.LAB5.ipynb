```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, RobustScaler, StandardScaler
```

```python
df = pd.read_csv("wine_data.csv")
```

```python
X = df.copy()
print("Original Data:\n", X.head())
```

```python
minmax = MinMaxScaler()
df_minmax = pd.DataFrame(minmax.fit_transform(X), columns=X.columns)

print("\nMin-Max Scaling:\n", df_minmax.head())
```

```python
df_mean_norm = (X - X.mean())/(X.max() - X.min())

print("\nMean Normalization:\n", df_mean_norm.head())
```

```python
maxabs = MaxAbsScaler()
df_maxabs = pd.DataFrame(maxabs.fit_transform(X), columns=X.columns)

print("\nMax-Abs Scaling:\n", df_maxabs.head())
```

```python
robust = RobustScaler()
df_robust = pd.DataFrame(robust.fit_transform(X), columns=X.columns)

print("\nRobust Scaling:\n", df_robust.head())
```

```python
standard = StandardScaler()
df_standard = pd.DataFrame(standard.fit_transform(X), columns=X.columns)

print("\nStandardization (Z-score):\n", df_standard.head())
```
